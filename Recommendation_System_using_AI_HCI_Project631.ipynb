```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Define Pet Attributes and User Preferences
pet_attributes = {
    'Size': ['Small', 'Medium', 'Large'],
    'Activity Level': ['Low', 'Moderate', 'High'],
    'Temperament': ['Calm', 'Friendly', 'Energetic', 'Independent'],
    'Grooming Needs': ['Low', 'Moderate', 'High'],
    'Compatibility with Children': ['Yes', 'No', 'Depends'],
    'Compatibility with Other Pets': ['Yes', 'No', 'Depends'],
    'Aggression Level': ['Low', 'Moderate', 'High']
}

user_preferences = {
    'Size': None,
    'Activity Level': None,
    'Temperament': None,
    'Grooming Needs': None,
    'Compatibility with Children': None,
    'Compatibility with Other Pets': None,
    'Aggression Level': None
}

# 2. Create Sample Pet Data
sample_pets_data = [
    {
        'Name': 'Golden Retriever',
        'Size': 'Large',
        'Activity Level': 'High',
        'Temperament': 'Friendly',
        'Grooming Needs': 'Moderate',
        'Compatibility with Children': 'Yes',
        'Compatibility with Other Pets': 'Yes',
        'Aggression Level': 'Low'
    },
    {
        'Name': 'Siamese Cat',
        'Size': 'Small',
        'Activity Level': 'Moderate',
        'Temperament': 'Energetic',
        'Grooming Needs': 'Low',
        'Compatibility with Children': 'Depends',
        'Compatibility with Other Pets': 'Depends',
        'Aggression Level': 'Moderate'
    },
    {
        'Name': 'Hamster',
        'Size': 'Small',
        'Activity Level': 'High',
        'Temperament': 'Independent',
        'Grooming Needs': 'Low',
        'Compatibility with Children': 'No',
        'Compatibility with Other Pets': 'No',
        'Aggression Level': 'Low'
    },
    {
        'Name': 'Turtle',
        'Size': 'Small',
        'Activity Level': 'Low',
        'Temperament': 'Calm',
        'Grooming Needs': 'Low',
        'Compatibility with Children': 'Yes',
        'Compatibility with Other Pets': 'Yes',
        'Aggression Level': 'Low'
    },
    {
        'Name': 'Border Collie',
        'Size': 'Medium',
        'Activity Level': 'High',
        'Temperament': 'Energetic',
        'Grooming Needs': 'Moderate',
        'Compatibility with Children': 'Yes',
        'Compatibility with Other Pets': 'Yes',
        'Aggression Level': 'Moderate'
    },
    {
        'Name': 'Sphynx Cat',
        'Size': 'Small',
        'Activity Level': 'Moderate',
        'Temperament': 'Friendly',
        'Grooming Needs': 'High',
        'Compatibility with Children': 'Yes',
        'Compatibility with Other Pets': 'Yes',
        'Aggression Level': 'Low'
    },
    {
        'Name': 'Pug',
        'Size': 'Small',
        'Activity Level': 'Low',
        'Temperament': 'Friendly',
        'Grooming Needs': 'Low',
        'Compatibility with Children': 'Yes',
        'Compatibility with Other Pets': 'Yes',
        'Aggression Level': 'Low'
    },
    {
        'Name': 'Parrot',
        'Size': 'Small',
        'Activity Level': 'High',
        'Temperament': 'Energetic',
        'Grooming Needs': 'Moderate',
        'Compatibility with Children': 'Yes',
        'Compatibility with Other Pets': 'Depends',
        'Aggression Level': 'Low'
    }
]

# 3. Develop Recommendation Logic
def recommend_pets(user_preferences, sample_pets_data):
    recommended_pets = []

    for pet in sample_pets_data:
        # Always recommend Parrot
        if pet['Name'] == 'Parrot':
            recommended_pets.append(pet)
            continue # Skip preference matching for Parrot

        match = True
        for attribute, preferred_value in user_preferences.items():
            if preferred_value is not None and preferred_value != 'No preference':
                pet_value = pet.get(attribute)

                if preferred_value == 'Yes' or preferred_value == 'No':
                    if pet_value == 'Depends':
                        pass
                    elif pet_value != preferred_value:
                        match = False
                        break
                elif pet_value != preferred_value:
                    match = False
                    break
        if match:
            recommended_pets.append(pet)
    return recommended_pets

# 4. Design and Implement GUI Input
preference_widgets = {}

for attribute, values in pet_attributes.items():
    options = ['No preference'] + values
    dropdown = widgets.Dropdown(
        options=options,
        value='No preference',
        description=attribute,
        disabled=False,
    )
    preference_widgets[attribute] = dropdown

# Output widget for recommendations
recommendation_output = widgets.Output()

# 5. Integrate GUI with Recommendation Logic and Display Results
def on_preference_change_with_display(change):
    with recommendation_output:
        clear_output()
        current_user_preferences = {}
        for attribute, widget in preference_widgets.items():
            if widget.value == 'No preference':
                current_user_preferences[attribute] = None
            else:
                current_user_preferences[attribute] = widget.value

        latest_recommendations = recommend_pets(current_user_preferences, sample_pets_data)

        print("Current User Preferences:", current_user_preferences)
        print("Recommended Pets:")
        if latest_recommendations:
            for pet in latest_recommendations:
                print(f"- {pet['Name']}")
        else:
            print("- No pets recommended for these preferences.")

# Attach observer to all widgets
for attribute, widget in preference_widgets.items():
    widget.observe(on_preference_change_with_display, names='value')

# Create and display GUI layout
gui_layout = widgets.VBox([
    widgets.HTML('<h3 style="color: #4CAF50;">Select your pet preferences:</h3>'),
    *[widget for widget in preference_widgets.values()],
    widgets.HTML('<hr style="border-top: 1px solid #ddd;">'),
    widgets.HTML('<h3 style="color: #2196F3;">Recommendations:</h3>'),
    recommendation_output
])

display(gui_layout)

# Trigger an initial recommendation to populate the output widget
on_preference_change_with_display(None)

print("\nPet recommendation system GUI initialized and ready!")
```
