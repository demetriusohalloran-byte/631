```python
# simple_chatbot.py
# Simple Traditional Rule-Based Chatbot

import datetime
import requests  # For future Azure extension (currently inactive)

def list_capabilities():
    """Return formatted list of chatbot capabilities."""
    capabilities = [
        "Greet you (say 'hello' or 'hi')",
        "Tell the current time (say 'time' or 'what time is it?')",
        "Answer 'how are you?'",
        "Show this list (type 'capabilities')",
        "Handle unknown inputs gracefully",
        "Exit cleanly (type 'exit')",
        "Extendable to Azure AI (see ask_azure_ai() function)"
    ]
    return "I can:\n- " + "\n- ".join(capabilities)

# === FUTURE EXTENSION: Azure Cognitive Services (QnA Maker) ===
def ask_azure_ai(question):
    """
    Connects to Azure Cognitive Services QnA Maker.
    Replace placeholders with real credentials to activate.
    """
    endpoint = "https://YOUR-RESOURCE.cognitiveservices.azure.com/language/:query-knowledgebases"
    key = "YOUR_SUBSCRIPTION_KEY"
    kb_id = "YOUR_KB_ID"

    url = f"{endpoint}{kb_id}/generateAnswer"
    headers = {"Ocp-Apim-Subscription-Key": key}
    body = {"question": question}

    try:
        response = requests.post(url, json=body, headers=headers, timeout=10)
        if response.status_code == 200:
            return response.json()["answers"][0]["answer"]
        else:
            return f"Azure error: {response.status_code}"
    except Exception as e:
        return f"Connection failed: {str(e)}"

# === MAIN CHAT LOOP ===
print("Welcome to Simple Chatbot!")
print("Type 'capabilities' to see what I can do, or 'exit' to quit.\n")

while True:
    try:
        user_input = input("You: ").strip().lower()

        if not user_input:
            print("Bot: Please say something!\n")
            continue

        if user_input == "exit":
            print("Bot: Goodbye! Have a great day!\n")
            break

        elif "capabilities" in user_input:
            print("Bot:", list_capabilities(), "\n")

        elif "hello" in user_input or "hi" in user_input:
            print("Bot: Hello! How can I help you today?\n")

        elif "how are you" in user_input:
            print("Bot: I'm doing great, thank you! I'm here to help.\n")

        elif "time" in user_input:
            current_time = datetime.datetime.now().strftime("%I:%M:%S %p")
            print(f"Bot: The current time is {current_time}.\n")

        else:
            # UNKNOWN INPUT â†’ OPTIONALLY USE AZURE AI
            print("Bot: I don't know that. Asking Azure AI (demo mode)...")
            ai_response = ask_azure_ai(user_input)
            print(f"Bot (Azure): {ai_response}\n")

    except (KeyboardInterrupt, EOFError):
        print("\nBot: Goodbye!")
        break

print("Chat ended.")
```
