````
import ipywidgets as widgets
from IPython.display import display, HTML

def generate_response(question):
    """Provides enhanced wine selection advice based on the question."""
    question_lower = question.lower()

    if not question.strip():
        return "Please ask a question about wine selection. You can ask about red wine, white wine, sweet/dry wines, or food pairings for chicken, beef, fish, or pasta."

    # --- Combination Queries (most specific should come first) ---
    # Dry red for chicken
    if "dry" in question_lower and "red" in question_lower and "chicken" in question_lower:
        return "For dry red wine with chicken, a light-bodied Pinot Noir or a Beaujolais would be excellent choices, especially for roasted or grilled chicken."
    # Light red for chicken
    if "light" in question_lower and "red" in question_lower and "chicken" in question_lower:
        return "A light red wine like Pinot Noir or a Gamay (Beaujolais) pairs wonderfully with chicken, especially roasted or grilled preparations."
    # Red for beef/steak
    if "red" in question_lower and ("beef" in question_lower or "steak" in question_lower):
        return "For red wine with beef or steak, a bold Cabernet Sauvignon or a rich Merlot are classic. For richer cuts, consider a Zinfandel or Syrah. If you like a more earthy touch, a Rioja or Sangiovese could be good."
    # White for fish/seafood
    if "white" in question_lower and ("fish" in question_lower or "seafood" in question_lower):
        return "For white wine with fish or seafood, a crisp Sauvignon Blanc or a Pinot Grigio works well with lighter dishes. For richer fish, an oaked Chardonnay or a Viognier can be superb. Shellfish often sings with a Muscadet or Albariño."
    # Dry white for fish/seafood
    if ("dry" in question_lower and "white" in question_lower and ("fish" in question_lower or "seafood" in question_lower)):
        return "A dry white wine like a crisp Sauvignon Blanc (e.g., Sancerre) or an un-oaked Chardonnay (e.g., Chablis) are excellent with most fish and seafood. Pinot Grigio is also a good dry choice."
    # Red for pasta with tomato sauce
    if "red" in question_lower and "pasta" in question_lower and ("tomato" in question_lower or "marinara" in question_lower):
        return "For pasta with tomato-based sauces, a medium-bodied red like Sangiovese (Chianti) or a Zinfandel will complement the acidity nicely. A Montepulciano d'Abruzzo is also a great option."
    # White for pasta with cream/pesto
    if "white" in question_lower and "pasta" in question_lower and ("cream" in question_lower or "pesto" in question_lower):
        return "For pasta with creamy sauces or pesto, a rich white wine like an oaked Chardonnay or a Vermentino would be delightful. A Pinot Grigio can work for lighter cream sauces."

    # --- Specific Food Pairings ---
    # Chicken
    if "chicken" in question_lower:
        if "grilled" in question_lower or "roasted" in question_lower:
            return "For grilled or roasted chicken, a versatile wine like Chardonnay (lightly oaked), Pinot Noir, or even a dry Rosé works well."
        elif "creamy" in question_lower or "rich" in question_lower:
            return "For creamy or rich chicken dishes, a full-bodied Chardonnay or a Viognier can stand up to the flavors. A lighter red like Pinot Noir can also work."
        elif "spicy" in question_lower:
            return "Spicy chicken pairs well with off-dry Riesling, Gewürztraminer, or a sparkling wine to balance the heat."
        else:
            # General chicken pairing, if no specific type is mentioned
            return "Chicken generally pairs well with a Chardonnay, Pinot Grigio, or a light-bodied Pinot Noir. For richer preparations, consider an oaked Chardonnay or Viognier. For lighter, try Sauvignon Blanc."

    # Beef/Steak
    if "beef" in question_lower or "steak" in question_lower:
        if "lean" in question_lower or "tenderloin" in question_lower:
            return "For leaner cuts of beef like tenderloin, a Merlot, Cabernet Franc, or a lighter Cabernet Sauvignon can be very enjoyable."
        elif "rich" in question_lower or "fatty" in question_lower or "ribeye" in question_lower:
            return "For rich or fatty cuts like ribeye, a bold Cabernet Sauvignon, Syrah, Zinfandel, or a Malbec are excellent choices to cut through the richness."
        else:
            # General beef/steak pairing
            return "For beef or steak, a Cabernet Sauvignon, Merlot, or Zinfandel would be excellent. Consider Syrah or Malbec for robust flavors."

    # Fish/Seafood
    if "fish" in question_lower or "seafood" in question_lower:
        if "light white fish" in question_lower or "cod" in question_lower or "tilapia" in question_lower:
            return "For light white fish, a crisp Sauvignon Blanc, Pinot Grigio, or Albariño are fantastic choices."
        elif "rich fish" in question_lower or "salmon" in question_lower or "tuna" in question_lower:
            return "For richer fish like salmon or tuna, an oaked Chardonnay, Viognier, or even a light-bodied Pinot Noir can be surprisingly good."
        elif "shellfish" in question_lower or "shrimp" in question_lower or "oysters" in question_lower:
            return "Shellfish, especially oysters, often shine with Muscadet, a dry sparkling wine, or a crisp Sauvignon Blanc. Shrimp can pair well with Pinot Grigio or un-oaked Chardonnay."
        else:
            # General fish/seafood pairing
            return "Most fish and seafood go well with a Sauvignon Blanc, Pinot Grigio, or a dry Riesling. For richer fish, you might enjoy a light Chardonnay or even a Pinot Noir."

    # Pasta
    if "pasta" in question_lower:
        if "tomato" in question_lower or "marinara" in question_lower:
            return "For pasta with tomato-based sauces, a medium-bodied red like Sangiovese (Chianti), Barbera, or a Zinfandel will complement the acidity nicely."
        elif "cream" in question_lower or "alfredo" in question_lower:
            return "For pasta with creamy sauces, a rich white wine like an oaked Chardonnay, Gavi, or a Pinot Grigio can be delightful. A light, unoaked Chardonnay or Vermentino also works."
        elif "pesto" in question_lower:
            return "Pesto pasta pairs wonderfully with a herbaceous white wine such as Sauvignon Blanc or Vermentino. Pinot Grigio is also a good choice."
        else:
            # General pasta pairing
            return "Pasta pairings depend on the sauce! For tomato-based, try Sangiovese. For creamy, a Chardonnay. For pesto, Sauvignon Blanc."

    # --- General Wine Type Queries ---
    if "red wine" in question_lower or "reds" in question_lower:
        if "bold" in question_lower or "full-bodied" in question_lower:
            return "For a bold or full-bodied red, I'd suggest a Cabernet Sauvignon, Syrah (Shiraz), Zinfandel, or Malbec. Look for regions like Bordeaux or Napa Valley."
        elif "light" in question_lower or "light-bodied" in question_lower:
            return "If you prefer a lighter red, try a Pinot Noir, Gamay (Beaujolais), or a Grenache. Italian options like a lighter Sangiovese can also be good."
        else:
            # General red wine suggestion
            return "For red wines, popular choices include Cabernet Sauvignon, Merlot, Pinot Noir, Syrah, Zinfandel, Malbec, and Sangiovese. You might explore regions like Bordeaux or Burgundy."

    if "white wine" in question_lower or "whites" in question_lower:
        if "crisp" in question_lower or "unoaked" in question_lower:
            return "A crisp, unoaked white wine like Sauvignon Blanc (e.g., Sancerre), Pinot Grigio, or an Albariño would be great choices."
        elif "rich" in question_lower or "oaked" in question_lower:
            return "For a rich or oaked white wine, a Chardonnay (e.g., from Burgundy or California) or a Viognier can be excellent."
        elif "sweet" in question_lower:
            return "For a sweet white, a Riesling (especially late-harvest) or a Moscato d'Asti would be lovely."
        else:
            # General white wine suggestion
            return "A crisp Sauvignon Blanc or a rich Chardonnay would be great choices for a white wine. Other options include Pinot Grigio, Riesling, Chenin Blanc, or Albariño. Explore regions like Sancerre, Chablis, or Alsace."

    if "sweet wine" in question_lower:
        return "If you like sweet wines, consider a Moscato d'Asti, a late-harvest Riesling, Port, Sauternes, or even an Ice Wine."

    if "dry wine" in question_lower:
        return "Many red and white wines are dry. For dry wines, you can look for a Sauvignon Blanc, Pinot Grigio, an un-oaked Chardonnay, Cabernet Sauvignon, Merlot, or most Pinot Noirs."

    if "pairing" in question_lower:
        return "What food are you looking to pair with wine? I can suggest options for chicken, beef, fish, or pasta. You can also ask for specific types of wine like 'red' or 'dry white'."

    # --- General Greetings (moved to end to prevent early triggering) ---
    if "hello" in question_lower or "hi" in question_lower:
        return "Hi there! I'm your enhanced wine selection bot. How can I help you choose a wine today?"

    # --- Fallback for unrecognized queries ---
    return f"I'm sorry, I can only provide specific wine suggestions based on type (red, white, sweet, dry) or food pairings (chicken, beef, fish, pasta). Your question about: '{question}' is a bit outside my current knowledge. Could you rephrase or ask about one of those categories?"

# Create the interactive widgets
question_input = widgets.Text(
    value='',
    placeholder='Type your wine question here (e.g., "red wine for chicken")',
    description='Question:',
    disabled=False,
    layout=widgets.Layout(width='80%')
)

response_output = widgets.Output()

def on_button_click(b):
    with response_output:
        response_output.clear_output()
        user_question = question_input.value
        bot_response = generate_response(user_question)
        print(f"Bot: {bot_response}")
        question_input.value = '' # Clear the input after submission

submit_button = widgets.Button(description="Get Wine Suggestion")
submit_button.on_click(on_button_click)

# Display the widgets
display(HTML("<h3>Wine Selection Bot (Interactive in Colab)</h3>"))
display(question_input, submit_button, response_output)
````

A `TclError: no display name and no $DISPLAY environment variable` error occurs when attempting to run Tkinter in Google Colab because Colab is a **cloud-based, headless environment**. This means it does not have a graphical display server (like X11 on Linux or the native display manager on Windows/macOS) that is necessary for GUI toolkits like Tkinter to render windows and widgets.

Tkinter, being a desktop GUI library, expects a visual display and an active display server to connect to. In a headless environment like Colab, such a server is absent. Therefore, when the `tk.Tk()` constructor is called, it cannot find a display to create a window on, leading to the observed `TclError`. To run Tkinter applications, you need a local Python environment with an active graphical display.
